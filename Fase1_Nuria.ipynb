```python
import pandas as pd
import warnings 
import datetime 

warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', None)
```

```python
for column in df.columns.tolist():
    print(column.upper(),':')
    print(df[column].dtype)
    if df[column].dtype == float or df[column].dtype == int:
        print('Max:', df[column].max()) 
        print('Min:', df[column].min())    
    print('Null:',df[column].isnull().sum())
    print(df[column].unique(),'\n')

```

```python
def to_abs(data, lista):
    for column in lista:
        data[column]=data[column].abs()

def to_int(data, lista):
    for column in lista:
        data[column]=data[column].round(0) 
        data[column]=data[column].astype('Int64')

def to_bool(data, lista):
    for column in lista:
        data[column]=data[column].astype(bool)

def to_date(data, lista):
    for column in lista:
        data[column]=pd.to_datetime(data[column])
        data[column]=data[column].dt.strftime('%Y-%m-%d %H:%M:%S')


```

```python
to_abs(df, ['adr'])

lista_decimales=df.select_dtypes(include=['float']).columns.tolist()
to_int(df, lista_decimales)

to_bool(df, ['is_canceled'])

# to_date(df, ['reservation_status_date'])
```
