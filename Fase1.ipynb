```python
import pandas as pd
import warnings 
import datetime
import numpy as np
import seaborn as sns

warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', None)
```

```python
url_finanzas_original = 'https://raw.githubusercontent.com/nuriancg/project-da-promo-H-modulo-4-team-Hotel_CSI/master/finanzas-hotel-bookings.csv'
df = pd.read_csv(url_finanzas_original, index_col=0)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
def to_abs(data, lista):
    for column in lista:
        data[column]=data[column].abs()

def to_int(data, lista):
    for column in lista:
        data[column]=data[column].round(0) 
        data[column]=data[column].astype('Int64')

def to_bool(data, lista):
    for column in lista:
        data[column]=data[column].astype(bool)

def to_date(data, lista):
    for column in lista:
        # Convertimos a string para que todas las fechas sean tratadas como cadenas
        data[column]=data[column].astype(str)
        # Normalizamos las fechas, agregando la hora por defecto
        data[column]=data[column].apply(lambda x: x if ' ' in x else x + '00:00:00')
        # Convertimos las fechas a datetime. Las fechas inválidas se convertiran a NaT
        # Manejamos los errores con el parametro errors='coerce' --> los valores invalidos o fuera de rango se convertiran a NaT
        data[column]=pd.to_datetime(data[column], errors='coerce')
        # Modificamos los nulos
        data[column]=data[column].fillna('NaT')

def to_undefined (data, lista):
    for column in lista:
        data[column]=data[column].replace(np.nan, 'Undefined')

```

```python
to_abs(df, ['adr'])

lista_decimales=df.select_dtypes(include=['float']).columns.tolist()
to_int(df, lista_decimales)

to_bool(df, ['is_canceled'])

to_date(df, ['reservation_status_date'])
```

```python
lista_col_to_undefined = ['market_segment','distribution_channel','country','market_segment','reservation_status','reserved_room_type','customer_type' ]
to_undefined(df, lista_col_to_undefined)
```

```python
dict_meses={'1':'January',
           '2':'February',
           '3':'March'}

df['arrival_date_month']=df['arrival_date_month'].replace(dict_meses)
```

```python
def to_moda(data, lista):
    for column in lista: 
        moda= data[column].mode()[0]
        data[column]=data[column].fillna(moda)

def to_mediana(data, lista):
    for column in lista: 
        data[column]=data[column].fillna(data[column].median().astype(int))
```

```python
lista_col_to_moda = ['hotel','distribution_channel','is_repeated_guest','assigned_room_type','agent']
lista_col_to_mediana = ['lead_time','adults','children','babies','previous_cancellations','previous_bookings_not_canceled','booking_changes','days_in_waiting_list','adr','required_car_parking_spaces','total_of_special_requests']

to_moda(df, lista_col_to_moda)
to_mediana(df, lista_col_to_mediana)
```

```python
dict_meal={np.nan:'SC',
           'Undefined':'SC'}

df['meal']=df['meal'].replace(dict_meal)
```
